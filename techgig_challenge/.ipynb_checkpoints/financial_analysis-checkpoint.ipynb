```python
axis_bnk_df.head()
```

```python
axis_bnk_df['returns'] = axis_bnk_df['Close'].pct_change(1)
```

```python
sbi_bnk_df['returns'] = sbi_bnk_df['Close'].pct_change(1)
indus_bnk_df['returns'] = indus_bnk_df['Close'].pct_change(1)
yes_bnk_df['returns'] = yes_bnk_df['Close'].pct_change(1)
kotak_bnk_df['returns'] = kotak_bnk_df['Close'].pct_change(1)
icici_bnk_df['returns'] = icici_bnk_df['Close'].pct_change(1)
hdfc_bnk_df['returns'] = hdfc_bnk_df['Close'].pct_change(1)
```

```python
sbi_bnk_df['returns'].hist(bins=100)
```

```python
axis_bnk_df['returns'].hist(bins=100,label='Axis', figsize=(16,10), alpha=0.4)
sbi_bnk_df['returns'].hist(bins=100,label='SBI', figsize=(16,10), alpha=0.4)
indus_bnk_df['returns'].hist(bins=100,label='Indus', figsize=(16,10), alpha=0.4)
yes_bnk_df['returns'].hist(bins=100,label='YES', figsize=(16,10), alpha=0.4)
kotak_bnk_df['returns'].hist(bins=100,label='Kotak', figsize=(16,10), alpha=0.4)
icici_bnk_df['returns'].hist(bins=100,label='ICICI', figsize=(16,10), alpha=0.4)
hdfc_bnk_df['returns'].hist(bins=100,label='HDFC', figsize=(16,10), alpha=0.4)
plt.legend();
```

```python
axis_bnk_df['returns'].plot(kind='kde',label='Axis', figsize=(16,10))
sbi_bnk_df['returns'].plot(kind='kde',label='SBI', figsize=(16,10))
indus_bnk_df['returns'].plot(kind='kde',label='Indus', figsize=(16,10))
yes_bnk_df['returns'].plot(kind='kde',label='YES', figsize=(16,10))
kotak_bnk_df['returns'].plot(kind='kde',label='Kotak', figsize=(16,10))
icici_bnk_df['returns'].plot(kind='kde',label='ICICI', figsize=(16,10))
hdfc_bnk_df['returns'].plot(kind='kde',label='HDFC', figsize=(16,10))
plt.legend();
```

## Cumulative Daily Return

```python
axis_bnk_df['cumulative_return'] = (1+axis_bnk_df['returns']).cumprod()
```

```python
sbi_bnk_df['cumulative_return'] = (1+sbi_bnk_df['returns']).cumprod()
indus_bnk_df['cumulative_return'] = (1+indus_bnk_df['returns']).cumprod()
yes_bnk_df['cumulative_return'] = (1+yes_bnk_df['returns']).cumprod()
kotak_bnk_df['cumulative_return'] = (1+kotak_bnk_df['returns']).cumprod()
icici_bnk_df['cumulative_return'] = (1+icici_bnk_df['returns']).cumprod()
hdfc_bnk_df['cumulative_return'] = (1+hdfc_bnk_df['returns']).cumprod()
```

```python
axis_bnk_df['cumulative_return'].plot(label='Axis', figsize=(16,10))
sbi_bnk_df['cumulative_return'].plot(label='SBI', figsize=(16,10))
indus_bnk_df['cumulative_return'].plot(label='Indus', figsize=(16,10))
yes_bnk_df['cumulative_return'].plot(label='YES', figsize=(16,10))
kotak_bnk_df['cumulative_return'].plot(label='Kotak', figsize=(16,10))
icici_bnk_df['cumulative_return'].plot(label='ICICI', figsize=(16,10))
hdfc_bnk_df['cumulative_return'].plot(label='HDFC', figsize=(16,10))
plt.legend();
```
